```python
total = 0
count = 0

for i in range(10):
    while True:
        num = int(input("Enter a number: "))

        if num == 0:
            break

        if num < 0:
            continue

        if num % 2 == 0:
            total += num

        count += 1
        break

    if num == 0:
        break

print("Final sum:", total)
```

```python
num = int(input("Enter a number: "))

for i in range(1, 13):
    product = num * i

    if product % 3 == 0:
        continue

    if product > 50:
        break

    print(num, "x", i, "=", product)
```

```python
import random

secret_number = random.randint(1, 20)
attempts = 0

while True:
    guess = int(input("Guess the number (1-20): "))

    if guess < 1 or guess > 20:
        print("Invalid guess! Enter a number between 1 and 20.")
        continue

    attempts += 1

    if guess == secret_number:
        print("Congratulations! You guessed it in", attempts, "attempts.")
        break
    else:
        print("Wrong guess, try again.")
```

```python
n = int(input("Enter a number n: "))
count = 0

for i in range(1, n+1):
    if i > 50:
        break

    if i % 8 == 0:
        continue

    if i % 4 == 0:
        count += 1

print("Count of numbers divisible by 4 (skipping multiples of 8, stopping >50):", count)
```

```python
for i in range(1, 6):
    for j in range(1, i+1):
        if j == 3:
            continue
        if i * j > 6:
            break
        print(j, end=" ")
    print()
```
